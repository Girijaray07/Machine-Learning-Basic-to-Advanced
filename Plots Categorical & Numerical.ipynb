```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

**a) Scatter Plot (Numerical - Numerical)**

```python
sns.scatterplot(tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size")
```

**b) Bar Plot (Categorial - Numerical)**

```python
sns.barplot(tips, x="day", y="size", hue="sex")
```

**c) Box Plot (Categorical - Numerical)**

```python
sns.boxplot(data=tips, x="sex", y="tip", hue="time")
```

**d) Distplot Graph**

```python
sns.displot(data=iris, x="species", y="sepal_length")
```

**e) HeatMap (Categorical - Categorical)**

```python
sns.heatmap(pd.crosstab(tips["sex"], tips["day"]))
```

**f) Pairplot**

```python
sns.pairplot(iris, hue="species")
```

**g) Lineplot**
